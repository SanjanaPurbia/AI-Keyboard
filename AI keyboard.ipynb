```python
#importing libraries
import cv2
import cvzone
from cvzone.HandTrackingModule import HandDetector
from time import sleep
from pynput.keyboard import Controller
from selenium import webdriver
```

```python
cap= cv2.VideoCapture(0)
cap.set(3,1200)
cap.set(4,720)
```

```python
# #Opening web driver

driver_path = "D:\\Setup\\driver\\chromedriver_win32\\chromedriver.exe"
browser = webdriver.Chrome(executable_path=driver_path)
```

```python
detector=HandDetector(detectionCon=0.8)
```

```python
#drawing rectangle and text on it
def drawAll(img,buttonList):
    for button in buttonList:
        x,y=button.pos
        width,height=button.size
        cv2.rectangle(img, button.pos, (x + width, y + height), (255, 0, 255), cv2.FILLED)
        cv2.putText(img, button.text, (x + 10, y + 45), cv2.FONT_HERSHEY_PLAIN, 3, (255, 255, 255), 3)
        cvzone.cornerRect(img, (x, y, width, height), 10, rt=2)
    return img
```

```python
def search(string):
    if string !="":
        browser.get("https://www.google.com/")
        browser.find_element_by_class_name("gLFyf").send_keys(string)
        browser.find_element_by_xpath("/html/body/div[1]/div[3]/form/div[1]/div[1]/div[3]/center/input[1]").click()
```

```python

buttonList=[]

#defing keys of keyboard
keys=[["Q","W","E","R","T","Y","U","I","O","P"],
      ["A","S","D","F","G","H","J","K","L","<"],
      ["Z","X","C","V","B","N","M"," ",".",">"]]
finalText=""
```

```python
keyboard=Controller()
```

```python
for i in range(len(keys)):
    for x,key in enumerate(keys[i]):
        buttonList.append(Button([80*x+50, 80*i+25], key))
```

```python
#handling webcam 
while True:
    success,img=cap.read()
    img=detector.findHands(img)
    lmList ,BboxInfo = detector.findPosition(img)
    ##for x in range(0,5):
        #buttonList.append(Button([100*x+25,100],"Q"))
    img=drawAll(img,buttonList)
    #img=mybutton1.draw(img)
    #img=mybutton2.draw(img)
    
    #### Checking the hand is present or not ###
    if lmList:
        for button in buttonList:
            x,y=button.pos
            width,height=button.size
            #### cheching our finger and tip is there point no 8(index finger) in mediapipe liberary ###
            ##lmList[8][0] checking x-axis tip is b/w x and x+w then its in correct range then we can change the button ###
            ## lmList[8][1] checking y-axis   ###
            if x<lmList[8][0]<x+width and y<lmList[8][1]<y+height:
                cv2.rectangle(img,button.pos,(x+width,y+height),(175,0,175),cv2.FILLED)
                cv2.putText(img,button.text,(x+10,y+45),cv2.FONT_HERSHEY_PLAIN,3,(255,255,255),3)
                
                l,_,_ = detector.findDistance(8,12,img,draw=False)
                print(l)
                
                if l<40:
                    keyboard.press(button.text)
                    cv2.rectangle(img, button.pos, (x + width, y + height), (0,255,0), cv2.FILLED)
                    cv2.putText(img, button.text, (x + 10, y + 45), cv2.FONT_HERSHEY_PLAIN, 3, (255, 255, 255), 3)
                    
                    if button.text==">":
                        search(finalText)
                        finalText = ""
                    elif button.text =="<":
                        finalText=finalText[:-1]
                    else:
                        finalText+=button.text   #adding text to empty string
                    sleep(0.15)
     ## making a rextangle to put text                
    cv2.rectangle(img,(60,350),(900,450),(255,0,255),cv2.FILLED)
    cv2.putText(img,finalText,(75,425),cv2.FONT_HERSHEY_PLAIN,3,(255,255,255),3)          
            
    cv2.imshow("Video",img)
    if cv2.waitKey(1)==13:
        break
    
cam.release()
cv2.destroyAllWindows()
```
